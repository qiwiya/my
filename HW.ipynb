```python
'''В классе каждый ученик должен выбрать себе два кружка из трёх. 
Ведущий каждого кружка принёс классному руководителю неупорядоченные имена учеников, записанным к нему. 
Каждый ученик выбрал два кружка, но некоторые ученики решили схитрить и записались на все кружки сразу. 
Вычислите, сколько их. Гарантируется, что в каждом кружке есть хотя бы один ученик.

ФОРМАТ ВВОДА
    На первой строке — список учеников в первом кружке через запятую и пробел; 
    на второй — список учеников во втором кружке через запятую и пробел; 
    на третьей — список учеников в третьем кружке через запятую и пробел. 
    
 ФОРМАТ ВЫВОДА
    Число, количество учеников, записанных сразу на три кружка'''

list1 = input().split(', ')
list2 = input().split(', ')
list3 = input().split(', ')

#list1 += list2 + list3
#print(list1)
i = 0
counter = 0

while i < len(list1):
    if list1[i] in list2:
        if list1[i] in list3:
            counter +=1
    i+=1

print(counter)
        
```

```python
'''Вася и Маша играют в игру: 
Вася пишет какое-то слово, а затем Маша начинает составлять слова из букв слова Васи. 
Составьте программу, которая проверяет для какого-либо слова, может ли Маша использовать его в игре. 
Буквы могут повторяться 
(например, если Вася написал "кошмар", Маша может написать "мама" — и "м", и "а" есть в исходном слове) 

ФОРМАТ ВВОДА
    На первой строке — слово Васи со строчной буквы; 
    на второй — слово от Маши.
    
ФОРМАТ ВЫВОДА
    Если слово Маши можно использовать: "ДА".
    Если слово нельзя использовать: "НЕТ".'''

wordVasya = list(input().lower())
wordMasha = list(input().lower())
result = ''
i = 0

while i < len(wordMasha):
    if wordMasha[i] in wordVasya:
        result = 'ДА'
        i +=1
    else:
        result = 'НЕТ'
        break
print(result)

```

```python
'''Леонид печет торты и ведет профессиональный аккаунт в Instagram. 
Он хочет собрать статистику, 
сколько уникальных пользователей оставили комментарии к его постам на прошлой неделе. 
Кроме того, у Леонида есть список спам-аккаунтов, 
и если комментарий оставлен с такого аккаунта, то Леонид его не учитывает. 
Напишите программу, которая поможет Леониду. 

ФОРМАТ ВВОДА
    Сначала вводятся аккаунты, с которых Леониду написали комментарии.
    Каждый аккаунт вводится на новой строке, когда комментарии заканчиваются, вводится строка "END". 
    Далее вводятся спам-аккаунты, среди них могут быть те, которые не оставляли на прошлой неделе комментариев. 
    Каждый аккаунт вводится на новой строке, когда комментарии заканчиваются, вводится слово "END". 

ФОРМАТ ВЫВОДА
    Целое число — количество уникальных не спам-аккаунтов, которые написали комментарии Леониду.'''

listAllAcc = set()
listSpamAcc = []
i = 0
counter = 0

while True:
    txt = input()
    if txt.lower() != 'end':
        listAllAcc.add(txt)
    else: break
        
while True:
    txt = input()
    if txt.lower() != 'end':
        listSpamAcc.append(txt)
    else: break
        
while i < len(listAllAcc):
    if list(listAllAcc)[i] not in listSpamAcc:
        counter +=1
        i +=1
    else: i+=1
print(counter)
```

```python
'''Олег считает своими счастливыми числами все, 
которые делятся на 4 и на 6 одновременно (например, 12). 
Олег хочет написать программу, которая принимает на ввод число и выводит 
все счастливые числа Олега в промежутке от 1 до этого числа включительно. 
Если таких чисел нет, то программа ничего не выводит.

ФОРМАТ ВВОДА 
    Целое положительное число N. 
    
ФОРМАТ ВЫВОДА 
    Все числа в промежутке от 1 до N включительно, 
    которые являются счастливыми для Олега, каждое с новой строки.'''

num = int(input())
i = 12

while i <= num:
    print(i)
    i +=12
```

```python
'''Группа "Багровый Фантомас" собирается записать альбом на виниловой пластинке, 
но пластинка не вместит материал дольше 45 минут. Помогите группе определить, сколько песен они смогут выпустить.

ФОРМАТ ВВОДА
    На каждой строке вводится длина песни в минутах и секундах в формате m:ss. 
    Гарантируется, что каждая песня длится не больше 9 минут 59 секунд.
    Ввод заканчивается, когда суммарная длительность песен превышает 45 минут.

ФОРМАТ ВЫВОДА
    Количество песен, которые уместятся на пластинке.'''

duration = 0
countSong = 0

while duration <= 45*60:
    duration1song = input()
    duration1song = int(duration1song[0])*60+int(duration1song[2:])
    duration += duration1song
    if duration <= 45*60:
        countSong +=1
    else: break
print(countSong)
```

```python
'''Повышенная стипендия за отличную учебу равна 20000. 
Вася решил посчитать, хватит ли ему денег на жизнь, если он получит стипендию. 
Он посчитал и записал свои расходы за последние 12 месяцев. 
Помогите Васе найти количество месяцев, когда ему явно не хватило бы стипендии.

ФОРМАТ ВВОДА
    12 целых неотрицательных чисел, каждое с новой строки.
    
ФОРМАТ ВЫВОДА
    Число, означающее количество месяцев, когда Вася тратил более 20 000 рублей в месяц.'''
grant = 20000
countmonth = 0
i = 0

while i < 12:
    expense = int(input())
    i +=1
    if expense > grant:
        countmonth += 1
print(countmonth)
```

```python
'''Вася каждый месяц придерживался разного режима питания, 
а потом решил выяснить, в какой месяц у него были самые дешевые обеды. 
Он посчитал и записал затраты на питание за последние несколько месяцев. 
Вычислите, какое наименьшее количество рублей в месяц Вася потратил на обеды.

ФОРМАТ ВВОДА
    Целые неотрицательные числа, не превосходящие 1 миллион, каждое с новой строки.
    Затем на новой строке вводится отрицательное число, обозначающее конец ввода. 
    
ФОРМАТ ВЫВОДА
    Число, равное наименьшему введенному числу (не считая последнего, означающего конец ввода).'''

expense = int(input())
minexpense = expense

while expense >= 0:
    expense = int(input())
    if expense < 0:
        break
    elif expense < minexpense:
        minexpense = expense
print(minexpense)
```

```python
'''Вася заметил, что с каждым месяцем у него все хуже получается откладывать деньги на новый ноутбук. 
Он подумал, что это может произойти по той причине, что с каждым месяцем он тратит все больше и больше. 
Проверьте, действительно ли в каждый следующий месяц Вася тратил больше, чем в предыдущий.

ФОРМАТ ВВОДА
    Целые неотрицательные числа, не превосходящие 1 миллион, каждое с новой строки. 
    Затем с новой строки — отрицательное число, обозначающее конец ввода. 
    
ФОРМАТ ВЫВОДА
    True, если в каждый следующий месяц Вася тратил больше, чем в предыдущий
    (без учета последнего введенного числа, означающего конец ввода), и False в противном случае. 
    Если были введены расходы всего за ноль или один месяц, ответ True.'''

expense = int(input())
prevexpense = expense
result = True

while expense >= 0:
    expense = int(input())
    if expense < 0:
        break
    elif expense > prevexpense:
        # print(expense, prevexpense, result)
        prevexpense = expense
    else: result = False
print(result)
```
